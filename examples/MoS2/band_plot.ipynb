```python
from dptb.nn.build import build_model
import torch
from dptb.postprocess.bandstructure.band import Band
import matplotlib.pyplot as plt 
import numpy as np
```

```python
model = build_model('./train/train_out/checkpoint/nnsk.best.pth')
```

```python
task_options={
        "task": "band",
        "kline_type":"abacus",
        "kpath":[
            [0, 0, 0, 30],
            [0.5, 0, 0, 30],
            [0.3333333, 0.3333333, 0, 30],
            [0, 0, 0, 1]
        ],
        "nel_atom":{"Mo":6,"S":6},
        "klabels":["G", "M", "K", "G"],
        "emin":-15,
        "emax":5,
        "ref_band": './train/data/kpath.0/eigenvalues.npy'
    }
kpath_kwargs = task_options

stru_data = "./MoS2_unit_rlx.vasp"



band = Band(model, results_path='./bandplot')
AtomicData_options = { "pbc": True}
band.get_bands(data = stru_data, 
               kpath_kwargs = task_options,
               AtomicData_options = AtomicData_options)
band.band_plot(emin = task_options['emin'],
               emax = task_options['emax'],
               ref_band = task_options['ref_band'],)
```
